### 3.3 Setting up the validation framework
____



* Perform the train/validation/test split with Scikit-Learn


- The df_train_full data is already split into train and test of 80% and 20% respectively/ 

- We will split the df_train_full data into train and validation.

```python
# importing the model for data splitting
from sklearn.model_selection import train_test_split
```

```python
# splitting the data into training and testing
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)


# splitting the training data into training and validation
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=11)
```

```python
# checking the length of the data
len(df_train), len(df_val), len(df_test)
```

```python
# reseting the index after splitting 
df_train_full = df_train_full.reset_index(drop=True)
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
# assigning the target variables to the y_train and y_val
y_train = df_train.churn.values
y_val = df_val.churn.values
y_test = df_test.churn.values
```

```python
# deleting the target variables from the data
del df_train['churn']
del df_val['churn']
del df_test['churn']
```

#### 3.8 One-hot encoding
____

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
# Initialize DictVectorizer
dv = DictVectorizer(sparse=False)


# Convert DataFrame to list of dictionaries
train_dict = df_train[categorical + numerical].to_dict(orient='records')
val_dict =  df_val[categorical + numerical].to_dict(orient='records')


# Transform the list of dictionaries into a matrix
X_train = dv.fit_transform(train_dict)
X_val = dv.transform(val_dict)


# # Get feature names
feature_names = dv.get_feature_names_out()

# Display the encoded matrix and feature names
print(pd.DataFrame(X_train, columns=feature_names))
```

```python
list(feature_names)
```

#### 3.9 Machine learning for classification: Logistic regression
____

```python
# importing the model
from sklearn.linear_model import LogisticRegression
```

```python
# fitting the model
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)
```

```python
# model intercept
model.intercept_[0]
```

```python
# model coef
model.coef_[0].round(3)
```

```python
# model prediction
y_pred = model.predict_proba(X_val)
y_pred
```

```python
# model prediction focusing on the soft prediction
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
# setting the threshold to 0.5
churn_decision = (y_pred >= 0.5)
```

```python
# checking the accuracy of the model
(y_val == churn_decision).mean()
```

```python
df_pred = pd.DataFrame()
df_pred['probability'] = y_pred
df_pred['prediction'] = churn_decision.astype(int)
df_pred['actual'] = y_val
```

```python
df_pred
```

```python
df_pred['correct'] = df_pred.prediction == df_pred.actual
df_pred
```
