```python
# clean the column headers 
def clean_column_header(df):
    df.columns = df.columns.astype(str)
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace( ' ', '_')
    return df

# calling the function 
clean_column_header(df)
```

```python
# checking the content of the changes to the columnn header
# this can be easily seen with the df being transpose
df.T
```

```python
# channging the column context which are object to uniform content
object_columns_content = df.select_dtypes(include='object')
object_columns_content
```

```python
# transforming those columns to uniform content 
def clean_column_content(df, column_name):
    df[column_name] = df[column_name].astype(str)
    df[column_name] = df[column_name].str.lower()
    df[column_name] = df[column_name].str.replace(' ', '_')
    return df

for col in object_columns_content:
    clean_categorical_data_for_use = clean_column_content(df, col)
    
df
    
    
```
