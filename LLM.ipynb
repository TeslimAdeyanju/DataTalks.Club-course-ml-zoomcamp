## 2.3 Using the (Byte Pair Enconding) BPE Tokenizer

```python
# from tiktoken import SimpleTokenizerV1
import importlib
import tiktoken

print("tiktoken version:", importlib.metadata.version("tiktoken"))
```

```python
# calling the get_encoding method with a parameter of "gpt2"
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
# this is a simple text that we want to encode
text = (
    "Hello, I am Teslim, and learning to become a data scientist? <|endoftext|> \
    It is a great journey, and I am enjoying it.")

# encoding the text, and assigning the result to the variable "integers"
integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})

# printing the result
print(integers)
```

```python
# decoding the integers, and assigning the result to the variable "decoded"
decoded = tokenizer.decode(integers)

# printing the result
print(decoded)
```
